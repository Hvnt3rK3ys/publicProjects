## 2Punto OJO  c.

```r
edad = c(22, 34, 29, 25, 30, 33, 31, 27, 25, 25)
tiempo = c(14.21, 10.36, 11.89, 13.81, 12.03,
           10.99, 12.48, 13.37, 12.29, 11.92)
sexo = c("M","H","H","M","M","H","M","M","H","H")
misDatos = data.frame(edad,tiempo,sexo)
misDatos

# ATN A SUJETOS (filas) 3 a 6
misDatos[3:6,]


misDatos[,1]

misDatos$edad

#ATN DATAFRAME only men
# and only women

hombres=subset(misDatos,sexo=="H")

hombres

mujeres=subset(misDatos,sexo=="M")

mujeres


#men older than

mayores=subset(misDatos,sexo=="H" & edad>30)

mayores

#men less than 30 years and have spoken by phone:

jov_habladores=subset(misDatos,sexo=="H" & edad<30 & tiempo>12)

jov_habladores

#people 25 o más 30 años 

extremos=subset(misDatos,edad<25|edad>30)

extremos

# 2 Punto L c
# Construir subconjunto women more than 25 and less than 50.
#subconjunto have spoken by phone Media

summary(misDatos)
mujeres_Mayores=subset(misDatos,sexo =="M"& (edad>25|edad<50))
mujeres_Mayores
mean(mujeres_Mayores$tiempo)

```
